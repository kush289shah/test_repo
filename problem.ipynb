```python
start_stations = []
end_stations = []
stations = []
filename = "201601-citibike-tripdata.csv"
with open(filename, 'r') as csvfile:
    filereader = csv.reader(csvfile, delimiter=',')
    header = next(filereader)
    for row in filereader:
        stations.append(row[4])
        stations.append(row[8])
        start_stations.append(row[4])
        end_stations.append(row[8])
        
#print(start_stations)
#print(end_stations)

start_counts = {}
end_counts = {}
uniquestations = set(stations)
#print(uniquestations)

for station in uniquestations:
    start_counts[station] = start_stations.count(station)
    end_counts[station] = end_stations.count(station)
    
#print("-----")
#print(start_counts)
#print(end_counts)
#print("-----")

balance_stations = {}
for station in uniquestations:
    balance_stations[station] = abs(end_counts[station] - start_counts[station])
    
#print("---------")
#print(balance_stations)
for key in balance_stations:
    print(key ,":", balance_stations[key],"\n")
#print("---------")
```
